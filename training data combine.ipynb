```python
import xml.etree.ElementTree as ET

import pandas as pd

def read_info(file_name):
	xml_data = open(file_name,'r').read()
	root = ET.XML(xml_data)
	data = []
	for child in root:
		data.append([subchild.text] for subchild in child)
	df = pd.DataFrame(data)
	return df
```
